```python
# B1
employees = ["Beth", "Fred", "Drew", "Alex", "Emma", "Chad"]
years     = [15,     11,     8,      5,     4,      2]

print(employees)
print(years)
```

```python
# C1

quantities = [84, 100, 126, 150, 186, 200, 216, 248]
```

```python
# C2

for q in quantities:
    if q <= 100:
        cost = q * 100
    elif q <= 200:
        cost = 100 * 100 + (q - 100) * 90
    else:
        cost = 100 * 100 + 100 * 90 + (q - 200) * 80

    print(f"The cost for an order of {q} computer monitors is ${cost}.")
```
